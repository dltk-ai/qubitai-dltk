### Image Classification using DLTK

> Image Classification is a fundamental task that attempts to comprehend an entire image as a whole. The goal is to classify the image by assigning it to a specific label. Typically, Image Classification refers to images in which only one object appears and is analyzed. 

In this notebook we will use DLTK to classify images using open source classifiers & 3rd Party AI Engines

```python
# If you are running this notebook on GOOGLE COLAB, uncomment below code & run this cell

# !git clone https://github.com/dltk-ai/qubitai-dltk.git
# %cd qubitai-dltk
```

#### Import necessary packages

```python
from PIL import Image


import sys
sys.path.append("../../")

import json
```

#### Read Image

```python
image_path = "./examples/data/image/image_classification_sample_2.jpg"
image = Image.open(image_path)
image
```

#### Use DLTK supported open source models to classify images

#### Intialize DLTK client

> <u>Parameters Descriptions</u>
>
> **API key**: `If authentication is enabled`, you need to provide a valid API key
>
> **base_url**: url where kong/base service provided by OpenDLTK is deployed.

```python
import dltk_ai
client = dltk_ai.DltkAiClient(base_url='http://localhost:8000')
```

<div class="alert alert-block alert-info">
<b>See Also:</b> 

 

1. [How to generate API key](https://docs.dltk.ai/qubitai-dltk/docs/build/html/getting_started/generateAPIkey.html)
2. [OpenDLTK deployment](http://docs.dltk.ai/qubitai-dltk/docs/build/html/getting_started/openDLTK_setup.html)

 

</div>

```python
image_classification = client.image_classification(image_path=image_path, top_n=4, image_classifiers=['tensorflow'])
```

```python
image_classification_output = client.check_cv_job(image_classification['task_id'])
image_classification_output
```

#### Classify Images using supported AI engines

<div class="alert alert-block alert-danger"> <b>Caution:</b> Run below cell, only if <b>Supported AI Engine Credentials</b> are updated in openDLTK configurations while installation </div>

```python
image_classification = client.image_classification(image_path=image_path, top_n=4, image_classifiers=['azure'])
image_classification
```

```python
image_classification_output = client.check_cv_job(image_classification['task_id'])
image_classification_output
```
